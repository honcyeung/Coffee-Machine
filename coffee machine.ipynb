```python
global menu
menu = {
    "espresso": {
        "ingredients": {
            "water": 50,
            "coffee": 18,
        },
        "cost": 1.5,
    },
    "latte": {
        "ingredients": {
            "water": 200,
            "milk": 150,
            "coffee": 24,
        },
        "cost": 2.5,
    },
    "cappuccino": {
        "ingredients": {
            "water": 250,
            "milk": 100,
            "coffee": 24,
        },
        "cost": 3.0,
    }
}

resources = {
        "water": 300, 
        "milk": 200, 
        "coffee": 100, 
        "money": 0
    }

unit = ['ml', 'ml', 'g']
coins = ['quarters', 'dimes', 'nickles', 'pennies']

def sufficient(decision, resources):
    
    decision = menu[decision]
    ingredient = decision['ingredients']
    
    for item in ingredient:
        if resources[item] >= ingredient[item]: 
            return True, item
        else: 
            return False, item

def coin(decision, amount):
    
    cost = menu[decision]['cost']
    if amount >= cost: 
        return (amount-cost), True
    else: 
        return None, False
        
def update_resources(decision):
    
    decision = menu[decision]
    for key, value in decision['ingredients'].items():
        resources[key] -= value
        if value < 0: 
            resources[key] = 0
    resources['money'] += decision['cost']
    
    return resources
    
off = False
while not off:
    
    decision = input('What would you like? Espresso, Latte, or Cappuccino? ').lower()
    
    if decision == 'off':
        off = True
    elif decision == 'report':
        count = 0
        for key, item in resources.items():
            if key == 'money': print(f'{key.title()}: ${item}')
            else:
                print(f'{key.title()}: {item}{unit[count]}')
                count += 1
    elif decision in ['espresso', 'latte', 'cappuccino']:
        make, item = sufficient(decision, resources)
        
        if make:
            print('Please insert coins.')
            total = 0
            for count, value in enumerate(coins):
                total += int(input(f'How many {coins[count]}? '))
            change, proceed = coin(decision, total)
        
            if proceed:
                print(f'\nHere is ${change:.2f} in change.')
                print(f'Here is your {decision.title()} ☕️. Enjoy!')
                resources = update_resources(decision)            
            else: 
                print("Sorry. There isn't enough money. Money refunded.")
        else:
            print(f"Sorry. There isn't enough {item}")
    else:
        print('Wrong input!')
```
